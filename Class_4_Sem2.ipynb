```python
X = average_curves_norm.copy()
X = np.array (X.T) # Create an array as ML needs array to run
```

```python
X
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.datasets import make_blobs
```

```python
inertia = []
silhouette_avg = []  # Renamed the list to silhouette_avg
for i in range(1, 15):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
    labels = kmeans.labels_
    centers = kmeans.cluster_centers_
    if i > 1:
        silhouette_avg.append(silhouette_score(X, labels))  # Append silhouette scores to the list

# Silhouette Score
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score  # Import silhouette_score from sklearn.metrics

plt.plot(range(1, 15), inertia)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()

plt.plot(range(2, 15), silhouette_avg)  # Use the corrected list
plt.title('Silhouette Score')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
k  =6
kmeans= KMeans(n_clusters=k, random_state=0).fit(X)
```

```python
for c in range (k):
    average_curves_norm.T[kmeans.labels_ ==c].T.plot(legend=False)
    plt.plot(kmeans.cluster_centers_[c], linewidth=8, color='orange')
    plt.title("clusters" +str(c))
```
