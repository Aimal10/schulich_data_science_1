```python
import pandas as pd


# Calculate New Doses Administered
df['new_doses'] = df['total_doses_administered'].diff(1)

# Calculate New Doses for Unvaccinated
df['new_doses_unvaccinated'] = df['total_individuals_at_least_one'].diff(1)

# Calculate New Doses for People with One Vaccine
df['new_doses_one_vaccine'] = (df['total_individuals_partially_vaccinated']
                               - df['total_individuals_at_least_one']).diff()

# Calculate New Doses for People with Two Vaccines
df['new_doses_two_vaccines'] = (df['total_individuals_fully_vaccinated']
                                - df['total_individuals_partially_vaccinated']).diff()

# Calculate New Doses for People with Three Vaccines
df['new_doses_three_vaccines'] = (df['total_individuals_3doses']
                                  - df['total_individuals_fully_vaccinated']).diff()

# Print or analyze the resulting DataFrame
print(df[['report_date', 'new_doses', 'new_doses_unvaccinated', 'new_doses_one_vaccine', 'new_doses_two_vaccines', 'new_doses_three_vaccines']])





```
